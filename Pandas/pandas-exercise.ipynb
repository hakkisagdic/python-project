# Sort a Dataframe

```python
# sort by columns

df_exams.sort_values(by='math score')
```

```python
# sort descending by columns

df_exams.sort_values(by='math score', ascending=False)
```

```python
# sort by multiple columns
df_exams.sort_values(by=['math score', 'reading score'], ascending=False)
```

```python
#  sort descending by multiple columns and update the dataframe
df_exams.sort_values(by=['math score', 'reading score'], ascending=False, inplace=True)
```

```python
df_exams
```

```python
# sort descending with a key function
df_exams.sort_values('race/ethnicity', ascending=True, key=lambda x:x.str.lower())
```

# pivot_table()

```python
df_sales = pd.read_excel("supermarket_sales.xlsx")
df_sales
```

```python
df_sales.pivot_table(index="Gender", aggfunc="sum")
```

```python
df_sales.pivot_table(index="Gender", values=["Quantity", "Total"], aggfunc="sum")
```

```python
df_sales.pivot_table(index="Gender", columns="Product line", values="Total", aggfunc="sum")
```
