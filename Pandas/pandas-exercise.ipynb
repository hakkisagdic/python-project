```python
import pandas as pd
```

```python
df_population_raw_data = pd.read_csv('population_total.csv')
```

```python
df_population_raw_data
```

```python
df_population_raw_data.dropna(inplace=True) #drop the rows with missing values and update the dataframe with 'inplace'
```

```python
df_pivot = df_population_raw_data.pivot(index="year", columns="country", values="population")
```

```python
df_selected_countries = df_pivot[['United States', 'United Kingdom', 'Japan','Turkey']]
```

```python
df_selected_countries
```

# Lineplot

```python
df_selected_countries.plot(kind='line', xlabel='Year', ylabel='Population', title='Population (1955-2020)', figsize=(10,5))
```

```python
# export the dataframe with the selected countries as an excel file
 
df_selected_countries.to_excel('population_total_selected_countries.xlsx')
```

# Barplot

```python
import matplotlib.pyplot as plt 
```

```python
df_selected_countries_2020.plot(kind='bar', color='orange', xlabel='Country', ylabel='Population', title='Population (2020)', figsize=(10,5))
```
